## NLP TECHNOLOGIES

## step 1 : LowerCasing

```python
text = "I LOVE YOU"

# lower case
lower_text = text.lower()
print(lower_text)

# title case
title_text = text.title()
print(title_text)

# upper case
upper_text = text.upper()
print(upper_text)
```

### step 2 : punctuations

remove symbols like .,!;:]

```python
import string

text = " Wow!!! This is amazing. NLP is great,isn't it?"

# remove punctuation
clean_text = text.translate(str.maketrans('', '', string.punctuation))

print(clean_text)
```

### step 3 : Tokenization

  split text into smaller units (words or sentences)

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt', quiet=True)
text = 'NLP is fascinating. It allows machines to understand human language.'
tokens = word_tokenize(text)
print(tokens)
```

### step 4 : Remove the Stop words
  remove common words that do not add the strong meaning

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
text = 'NLP is fascinating. It allows machines to understand human language.'
tokens = word_tokenize(text)
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

### step 5 : Stemming 
 -reduce the word to its root form using rulke base cutting 

```python
# stemming 
from nltk.stem import PorterStemmer
text = 'NLP is fascinating. It allows machines to understand human language.'
tokens = word_tokenize(text)
stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(word) for word in tokens]
print(stemmed_tokens)
```

### Step 6 : Lemmatization
  - convert word to its dictionary base form

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)

text = 'NLP is fascinating. It allows machines to understand human language.'

tokens = word_tokenize(text)

lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

print(lemmatized_tokens)
```

### step 7 : Remove number
 - remove numeric value if not requried

```python
import re
text = 'price:$ 100.00'

#remove leading/trailing whitespace

clean_text = text.strip()
print(clean_text)

# remove number

clean_text=re.sub(r'\d+','',clean_text)
print(clean_text)
```

###  step 8 :Remove URL
  - remove web links from the text

```python
import re

# link text

text=" check out my project at http://localhost:8888/notebooks/NLP_data_cleaning.ipynb#NLP-TECH for more info"

clean_text = re.sub(r'https?://\S+','',text)

print(clean_text)
```

### step 9 : remove the HTML Tag
 - remove the html markup

```python
# import library to remove HTML tags
import re

text = "<p>This is a paragraph of text.</p><a href='https://example.com'>Link</a>"

clean_text = re.sub(r'<.*?>', '', text)

print(clean_text)
```

### step 10 : Remove the Emojis
  - remove emojis from the text

```python
import re

# emoji  text
text = "NLP is fascinating.😍🥰😘😡 It allows machines to understand human language 😍🥰😘😡"
# remove emojis
clean_text = re.sub(r'[^\w\s,]','',text)
print(clean_text)
```
