 ### NLP TECHNOLOGIES
    

### step 1 : LowerCasing

```python
text = """OMG!!! 😱 I can't believe this product cost me $4999 in 2024!!! Visit NOW at https://www.fakewebsite123.com/deal?id=7788&utm_source=ads 😡😡
Honestly, I’ve NEVER seen such a baaad quality item before... It’s sooo bad, like really really BAD!!! The packaging was broken <div>damaged</div> & the manual says “Refer to page 45-A” but there’s no page 45A??!! 😂
Contact me at example_user123@gmail.com or call +91-9876543210 ASAP!!!
"""
# lower case
lower_text = text.lower()
print(lower_text)
# title case
title_text = text.title()
print(title_text)
# upper case
upper_text = text.upper()
print(upper_text)
```

### step 2 : Puntuations
 remove symbols like .,!;:]

```python
import string
clean_text = text.translate(str.maketrans('', '', string.punctuation))
print(clean_text)
```

### step 3 : Tokenization
 split text into smaller units (words or sentences)

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt', quiet=True)
tokens = word_tokenize(text)
print(tokens)
```

### step 4 : Remove the Stop words
 remove common words that do not add the strong meaning




```python
from nltk.corpus import stopwords
nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)
```

### step 5 : Stemming
 -reduce the word to its root form using rulke base cutting

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print(stemmed_tokens)
```

### Step 6 : Lemmatization
  -convert word to its dictionary base form

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet', quiet=True)
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_tokens)
```

### step 7 : Remove number
remove numeric value if not requried

```python
import re
text_no_numbers = re.sub(r'\d+(\.\d+)?', '', text)  
print(text_no_numbers)
```

### step 8 :Remove URL
remove web links from the text

```python
import re
text_no_urls = re.sub(r'http\S+|www\S+', '', text)
print(text_no_urls)
```

### step 9 : remove the HTML Tag
remove the html markup

```python
import re
text_no_html = re.sub(r'<.*?>', '', text)
print(text_no_html)
```

### step 10 : Remove the Emojis
remove emojis from the text

```python
import re
text_no_emoji = re.sub(r'[^\x00-\x7F]+', '', text)
print(text_no_emoji)
```
